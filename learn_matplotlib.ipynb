# Matplotlib Object Oriented Method

Introduction to the OOM

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set x Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# Create Black Canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Main Picture')
 
# Larger Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Inset Picture')
```

subplots()

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=2, ncols=1)
```

```python
# Axes is an array of axes to plot on
axes
```

```python
# iterate through array
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# display the figure object
fig

# adjust automatically the position of the axes
```

```python
# adjust automatically the position of the axes
fig.tight_layout()
fig
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
```

```python
ax1 = axes[0][0]
ax1.plot(x, y, 'r--')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Gambar 1')

ax2 = axes[0][1]
ax2.plot(x, y, 'b-*')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Gambar 2')

ax3 = axes[1][0]
ax3.plot(x, y, 'g*-')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('Gambar 3')

ax4 = axes[1][1]
ax4.plot(x*2, (y*0.5)+10, 'm*', label='expense')
ax4.set_xlabel('holidays')
ax4.set_ylabel('expenses in million rupiah')
ax4.set_title('holidays vs expenses')
ax4.legend(loc=0)

plt.tight_layout()
fig
```

Figure size, aspect ratio, and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.
-figsize is a tuple of the width and height of the figure in inches

-dpi is the dots-per-inch (pixel per inch).


```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12,5))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

Saving Figure

Legends, labels, and titles

Figure titles

```python
fig, axes = plt.subplots(figsize=(12,5))

axes.plot(x, y, 'r', label= 'y')
axes.plot(x, y + 2, 'b--', label= 'y + 2')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
axes.legend()
```
